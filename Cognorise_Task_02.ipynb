```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Assuming X is a NumPy array
# Replace the following lines with your actual data loading or generation code
X = np.random.rand(4600, 12)  # Replace 10 with the actual number of features
y = np.random.rand(4600)

# Fit the linear regression model
lm = LinearRegression()
lm.fit(X, y)

# Convert X to a Pandas DataFrame for accessing columns
X_df = pd.DataFrame(X, columns=['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated'])

# Create a DataFrame with coefficients and feature names
coeff_df = pd.DataFrame(lm.coef_, X_df.columns, columns=['Coefficient'])

# Print the coefficient DataFrame
print(coeff_df)
```

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
sns.displot((y_test-predictions),bins=50)
```
